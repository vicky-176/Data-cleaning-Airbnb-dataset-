```python
import pandas as pd
```

```python
df = pd.read_csv('Airbnb.csv')
```

```python
df
```

```python
df.columns
```

```python
# 1
df = df.drop_duplicates(subset='id')
```

```python
df = df.dropna(subset=['id'])
```

```python
df
```

```python
# 2
df['NAME'] = df['NAME'].astype(str).str.strip()
```

```python
df = df.dropna(subset=['NAME'])
```

```python
df['NAME'] = df['NAME'].str.replace(r'[^\x00-\x7F]+', '', regex=True)
```

```python
df['NAME']
```

```python
# 3
df['host id'] = pd.to_numeric(df['host id'], errors='coerce')
```

```python
df = df.dropna(subset=['host id'])
```

```python
df['host id']
```

```python
# 4
df['host name'] = df['host name'].astype(str).str.strip()
```

```python
df = df.dropna(subset=['host name'])
```

```python
df['host name'] = df['host name'].str.replace(r'[^A-Za-z0-9 ]+', '', regex=True)
```

```python
df['host name']
```

```python
# 5
df = df.dropna(subset=['neighbourhood'])
```

```python
df['neighbourhood'] = df['neighbourhood'].str.title()
```

```python
df['neighbourhood']
```

```python
# 6
df = df.dropna(subset=['lat', 'long'])
```

```python
df = df[(df['lat'].between(-90, 90)) & (df['long'].between(-180, 180))]
```

```python
df[['lat','long']]
```

```python
# 7
df['country'] = df['country'].astype(str).str.strip()
```

```python
df['country code'] = df['country code'].astype(str).str.strip()
```

```python
df['instant_bookable'] = df['instant_bookable'].astype(str).str.strip()
```

```python
df['cancellation_policy'] = df['cancellation_policy'].astype(str).str.strip()
```

```python
df[['country','country code','instant_bookable','cancellation_policy']]
```

```python
# 8
df = df.dropna(subset=['room type'])
```

```python
df['room type'] = df['room type'].str.title()
```

```python
df['room type']
```

```python
# 9
df['Construction year'] = pd.to_numeric(df['Construction year'], errors='coerce')
```

```python
df['Construction year']
```

```python
df
```

```python
# 10
df['price'] = df['price'].replace('[\$,]', '', regex=True)
```

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
df = df.dropna(subset=['price'])
```

```python
df = df[df['price'] > 0]
```

```python
df['price']
```

```python
# 11
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True)
```

```python
df['service fee'] = pd.to_numeric(df['service fee'], errors='coerce')
```

```python
df['service fee']
```

```python
# 12
df['minimum nights'] = pd.to_numeric(df['minimum nights'], errors='coerce')
```

```python
df = df.dropna(subset=['minimum nights'])
```

```python
df = df[df['minimum nights'] > 0]
```

```python
df['minimum nights']
```

```python
# 13
df['number of reviews'] = pd.to_numeric(df['number of reviews'], errors='coerce')
```

```python
df = df.dropna(subset=['number of reviews'])
```

```python
df = df[df['number of reviews'] >= 0]
```

```python
df['number of reviews']
```

```python
# 14
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
```

```python
df = df.dropna(subset=['last review'])
```

```python
df['last review']
```

```python
# 15
df['availability 365'] = pd.to_numeric(df['availability 365'], errors='coerce')
```

```python
df = df.dropna(subset=['availability 365'])
```

```python
df = df[df['availability 365'].between(0, 365)]
```

```python
df['availability 365']
```

```python
# 16
df['reviews per month'] = pd.to_numeric(df['reviews per month'], errors='coerce')
```

```python
df = df.dropna(subset=['reviews per month'])
```

```python
df = df[df['reviews per month'] >= 0]
```

```python
df['reviews per month']
```

```python
df
```

```python
dp = df[cp]
```
